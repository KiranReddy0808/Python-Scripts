# Brownian Motion Experiment

Simulate discretized brownian path.


```python
import numpy as np
import matplotlib.pyplot as plt

T = 1
N = 500
dt = T/N

# Create arrays to store dW and W
dW = np.zeros(N)
W = np.zeros(N)

# Seed random number generator
np.random.seed(100)

# Generate dW and W
dW[0] = np.sqrt(dt)*np.random.randn()
W[0] = dW[0]
for i in range(1,N):
    dW[i] = np.sqrt(dt)*np.random.randn()
    W[i] = W[i-1] + dW[i]

# Create time array
time = np.linspace(0.0, N*dt, N)

# Create 2D histogram
hist, xedges, yedges = np.histogram2d(time, W, bins=30)

# Plot time vs W
plt.plot(time, W)
plt.xlabel('Time')
plt.ylabel('W')
plt.title('Plot of Time vs W')
plt.show()
```

Simulate a sample path of $\int_0^t W_s dW_s$ and compare it with the theoretical
result $\frac{1}{2}W_{t}^{2} - \frac{1}{2}t$.

```python
T = 3
N = 10000
dt = T/N
np.random.seed(100)
dW = np.sqrt(dt)*np.random.randn(N)
W = np.zeros(N)
Integral = np.zeros(N)
Exact = np.zeros(N)

for i in range(1, N):
    W[i] = W[i-1] + dW[i]
    Integral[i] = Integral[i-1] + W[i]*dW[i]
    Exact[i] = W[i]**2/2 - i*dt/2


plt.plot(Integral, Exact)
plt.xlabel('Integral')
plt.ylabel('Exact')
plt.title('Plot of Integral vs Exact')
plt.show()
```
